```python
adress_dataset = adress_dataset.assign(streetclass = '')
adress_dataset = adress_dataset.assign(street = '')
#adress_dataset.drop('street')
adress_dataset.head()
```

```python
classIterator = 0
for index, row in adress_dataset.iterrows():
    row['street'] = row['house_full_address'].split(',')[2]
    row['streetclass'] = classIterator
    if index<adress_dataset.shape[0]-1:
        if row['street'] != adress_dataset.iloc[index+1]['street']:
            classIterator+=1
    adress_dataset.iloc[index] = row
adress_dataset.head()
```

```python
x = adress_dataset.drop('streetclass',axis = 1)
x = x.drop('house_uuid',axis = 1)
y = adress_dataset['streetclass']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.10, stratify=y)
```
